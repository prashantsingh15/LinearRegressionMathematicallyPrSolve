```python
import pandas as pd
from sklearn.linear_model import LinearRegression


data = {
    'Square_Footage' : [1500, 2000, 1200, 1800, 2500],
    'Bedrooms': [3, 4 ,2 ,3 ,4],
    'Bathroom' : [2, 3, 1, 2, 3],
    'Price' : [250000, 350000, 200000, 300000, 400000]
}

df = pd.DataFrame(data)

X = df[['Square_Footage', 'Bedrooms', 'Bathroom']]
y = df['Price']

model = LinearRegression().fit(X, y)


print('Intercept', model.intercept_)
print('Coefficients:', model.coef_)


price_pred = 16700.4048 + 119.4331 * 1500 + 12651.82118 * 3 + 12651.8218 * 2
print("Sample Prediction For House:", price_pred)
```
